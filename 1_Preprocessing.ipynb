## 결합된 one-hot encoding

- 결혼여부 //
결혼상태 : 열 10개

- 경제활동 상태 //
변형근로시간 : 열 8개

- 성별 //
(만15세이상) 출산경험 /
생리여부 /
임신여부 : 열 8개

- 고혈압 //
검진당일 고혈압 약 복용 여부 : 열 3개 

- 당뇨병 //
검진당일 당뇨병 약 복용 여부: 열 3개 

```python
data['기혼|배우자|동거'] = (data['결혼여부']== 1) & (data['결혼상태']==1)
data['기혼|배우자|별거'] = (data['결혼여부']== 1) &(data['결혼상태']==2)
data['기혼|사별'] = (data['결혼여부']== 1) &(data['결혼상태']==3)
data['기혼|이혼'] = (data['결혼여부']== 1) &(data['결혼상태']==4)
data['기혼|응답거부'] = (data['결혼여부']== 1) &(data['결혼상태']==5)
data['기혼|모름'] = (data['결혼여부']== 1) &(data['결혼상태']==6)
```

```python
data.drop(columns = ['결혼상태'])
```

```python
data['직업|주간']	= (data['경제활동 상태']==1) & (data['변형근로시간']==1)
data['직업|오후']		= (data['경제활동 상태']==1) & (data['변형근로시간']==2)
data['직업|밤']		= (data['경제활동 상태']==1) & (data['변형근로시간']==3)
data['직업|주야교대']		= (data['경제활동 상태']==1) & (data['변형근로시간']==4)
data['직업|24교대']	= (data['경제활동 상태']==1) & (data['변형근로시간']==5)
data['직업|분할']	= (data['경제활동 상태']==1) & (data['변형근로시간']==6)
data['직업|불규칙']		= (data['경제활동 상태']==1) & (data['변형근로시간']==7)
data['직업|기타']	= (data['경제활동 상태']==1) & (data['변형근로시간']==8)
```

```python
data.drop(columns = ['변형근로시간'])
```

```python
data['여자|출산'] = (data['성별']== 2) & (data['(만15세이상) 출산경험']==1)
data['여자|출산x'] = (data['성별']== 2) & (data['(만15세이상) 출산경험']==1)
data['여자|생리'] = (data['성별']== 2) & (data['생리여부']==1)
data['여자|생리x|임신'] = (data['성별']== 2) & (data['생리여부']==0) & (data['임신여부']== 1)
data['여자|생리x|임신x'] = (data['성별']== 2) & (data['생리여부']==0) & (data['임신여부']== 0)
```

```python
data.drop(columns = ['(만15세이상) 출산경험','생리여부','임신여부'])
```

```python
data['당뇨병o|약복용o'] = (data['당뇨병']== 1) & (data['검진당일 당뇨병 약 복용 여부']==1)
data['당뇨병o|약복용x'] = (data['당뇨병']== 1) & (data['검진당일 당뇨병 약 복용 여부']==0)
data['당뇨병x|약복용x'] = (data['당뇨병']== 0) & (data['검진당일 당뇨병 약 복용 여부']==0)
```

```python
data.drop(columns = ['검진당일 당뇨병 약 복용 여부'])
```

```python
data['고혈압o|약복용o'] = (data['고혈압']==1) & (data['검진당일 고혈압 약 복용 여부']==1)
data['고혈압o|약복용x'] = (data['고혈압']==1) & (data['검진당일 고혈압 약 복용 여부']==0)
data['고혈압x|약복용x'] = (data['고혈압']==0) & (data['검진당일 고혈압 약 복용 여부']==0)
```

```python
data.drop(columns = ['검진당일 고혈압 약 복용 여부'])
```

```python
data = data.replace(False,0)
data = data.replace(True,1)
```
